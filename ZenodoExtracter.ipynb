```python
import requests
import os

def download_zenodo_data(record_id, destination_folder='.', filename=None):

    base_url = f"https://zenodo.org/api/records/{record_id}"

    try:
        response = requests.get(base_url)
        response.raise_for_status()
        data = response.json()

        files = data.get('files', [])
        if not files:
            print(f"No files found : {record_id}")
            return None


        download_file = files[0]

        file_url = download_file['links']['self']
        if not filename:
            filename = download_file['key']

        os.makedirs(destination_folder, exist_ok=True)
        local_filepath = os.path.join(destination_folder, filename)

        print(f"Downloading {filename} from {file_url}...")
        with requests.get(file_url, stream=True) as r:
            r.raise_for_status()
            with open(local_filepath, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        print(f"Download complete: {local_filepath}")
        return local_filepath


    except Exception as e:
        print(f"Error: {e}")
        return None

```

```python
path=download_zenodo_data("19509176", destination_folder='.', filename="data.zip")
```
